```python
f =open("DE.tmp")
l = f.readlines()
f.close()
```

```python
numVertices = int(l[0][:-1])
```

```python
s = 4727 
e = 42140
d = e-s+1
si = 2

numEdges = int(l[si+d+1][:-1])
```

```python
l[si+d+2:]
```

```python
import math

edges = []
b = l[si+d+2:]
a = []

for i in range(len(b)):
    if i%2 == 0:
        a.append(int(b[i].split(" ")[0]))
        a.append(int(b[i].split(" ")[1][:-1]))
    else:
        a.append(int(math.floor(float(b[i].split(" ")[1]))))
        edges.append(a)
        a = []
```

```python
edges
```

```python
f = open("temp_data_2.txt", "a")

f.write(str(numVertices)+" "+str(numEdges)+"\n")

for e in edges:
    f.write(str(" ".join([str(k) for k in e]))+"\n")

f.close()
```
